# Statistics and Visualization for Projects

## Load CSV File and Print Statistics

```python
import csv
import pandas as pd
        
projects = pd.read_csv('result/Projects.csv')

langs = ['JavaScript', 'Python', 'C', 'C++', 'Go', 'Java']
projects = projects[projects['language'].isin(langs)]
print('Total size: {}TB'.format(sum(projects['size']) / 1024 / 1024))
print('Number of projects: ', len(projects))
for lang in langs:
    print('Number of {} projects: {}'.format(lang, len(projects[projects['language'] == lang])))
projects.describe()
```

## Initialize Seaborn

```python
import seaborn as sns
sns.set(style='white')
sns.set(style='ticks')
```

## Visualize Distributions of Stars, Size, Commits, Authors, Watchers and Forks

```python
g = sns.catplot(data=projects, x='language', y='stars', kind='violin')
g.set(yscale='log')
```

```python
g = sns.distplot(projects['stars'], hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True))
```

## Visualize By Language

```python
import seaborn as sns
g = sns.FacetGrid(projects, col='language')
g.map(sns.distplot, 'stars', hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True))
```
